## Defining Own Function

```python
def hello_myName(name):
    print("Hello " + name + "!")
```

```python
hello_myName("Dawn")
```

```python
hello_myName("Sandy")
hello_myName("Hera")
```

## Functions with Default Value

```python
def hello(name ="stranger", hour =-1):
    
    print ("Hello " + name + "!")
    if hour == -1:
        print("Good Day!")
    else:
        if hour <= 1200:
            print("Good Morning1")
        elif hour <= 1900:
            print("Good Afternoon!")
        else:
            print("Good Evening!")
            
```

```python
hello()
```

```python
hello("Dawn")
```

```python
hello("Dawn", hour = 1400)
```

```python
hello("Dawn", hour = 1000)
```

```python
hello()
```

```python
def another_hello(name = "stranger"):
    print("Hello " + name + "!")
```

```python
another_hello("Dawn")
```

```python
another_hello()
```

## Functions with Return Value

```python
def area_of_rectangle(length, width):
    area = length * width
    return area
```

```python
area_1 = area_of_rectangle(4, 6)
area_1

```

```python
area_2 = area_of_rectangle(6, 10.5)
area_2
```
