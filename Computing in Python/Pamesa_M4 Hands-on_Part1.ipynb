Functions with Default Value

```python
def hello(name="stranger", hour=-1):
    
    print("Hello " + name + "!")
    
    if hour == -1:
        print("Good Day!")
    else:
        if hour <= 1200:
            print("Good Morning!")
        elif hour <= 1900:
            print("Good Afternoon!")
        else:
            print("Good Evening!") 
```

```python
hello()
```

```python
hello("Dawn")
```

```python
hello(name="Sandy", hour=1400)
```

```python
hello(hour=1000, name="Hera")
```

```python
hello(hour=2000)
```

## Minutes in a year calculator

```python
def calculate_minutes(days_in_a_year):
    hours_in_a_day = 24
    minutes_in_an_hour = 60
    
    total_minutes = days_in_a_year * hours_in_a_day * minutes_in_an_hour
    
    return total_minutes

standard_year = 365
minutes_standard = calculate_minutes(standard_year)
print(f"There are {minutes_standard} minutes in a standard {standard_year} year.")

leap_year = 366
minutes_leap = calculate_minutes(leap_year)
print(f"There are {minutes_leap} minutes in a leap {leap_year} year.")
```

## Radius of a Sphere Calculator

```python
import math

def calculate_circumference(radius):
    circumference = 2 * math.pi * radius
    return circumference

print("=== Sphere Circumference Calculator ===")
user_input = input("Enter the radius of the sphere: ")

radius = float(user_input)

result = calculate_circumference(radius)

print(f"The circumference of the sphere with radius {radius} is: {result:.2f}")
```
