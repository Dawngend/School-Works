## Activity 1

```python
def compute_average(score1, score2, score3):
    return (score1 + score2 + score3) / 3
```

```python
def get_grade(average):
    if average >= 90:
        return "A"
    elif average >= 80:
        return "B"
    elif average >= 70:
        return "C"
    else:
        return "F"
```

```python
def display_record(name, average, grade):

    print("\n===========================")
    print(f"Student: {name}")
    print(f"Average: {average:.2f}")
    print(f"Grade: {grade}")
    print("============================")

student_name = input("\nEnter Student Name:")

score1 = float(input("Enter Score 1:"))
score2 = float(input("Enter Score 2: "))
score3 = float(input("Enter Score 3: "))

calc_ave = compute_average(score1, score2, score3)
fin_grd = get_grade(calc_ave)

display_record(student_name, calc_ave, fin_grd)
```

## Activity 2

```python
def compute_subtotal(price, quantity):
    return price * quantity
```

```python
def apply_discount(subtotal, discount=0):
    return subtotal - discount
```

```python
def compute_total(subtotal, discount=0, tax_rate=0.12):
    disc_sub = apply_discount(subtotal, discount)
    tax_amt = disc_sub * tax_rate
    return disc_sub + tax_amt


item_price = float(input("\nEnter item price: "))
item_quantity = int(input("Enter item quantity: "))
discount_input = float(input("Enter discount amount: "))

subtotal_amt = compute_subtotal(item_price, item_quantity)

total_amt = compute_total(subtotal_amt, discount_input)

discounted_amt = apply_discount(subtotal_amt, discount_input)
tax_display = total_amt - discounted_amt

print(f"\nSubtotal: {subtotal_amt:.0f}")
print(f"Discount: {discount_input:.0f}")
print(f"Tax: {tax_display:.0f}")
print(f"Total: {total_amt:.0f}")
```

```python
def login(username, password):
    if username == "admin" and password == "1234":
        print("\nLogin Successful!")
        print(f"Welcome, {username}\n")
        return True
    else:
        print("\nLogin Failed!")
        print("Invalid username or password.\n")
        return False
```

```python
def check_access(is_logged_in):
    if is_logged_in:
        print("Access Granted")
    else:
        print("Access Denied")
```

```python
def logout(username):
    print("Logged out successfully.\n")

print("--- System Login ---")
entered_username = input("Enter username: ")
entered_password  = input("Enter password: ")

status = login(entered_username, entered_password)
check_access(status)

if status == True:
    print() 
    logout(entered_username)
```
